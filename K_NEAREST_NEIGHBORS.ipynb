```python
import pandas as pd
import numpy as np

# 1. Load dataset
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(url)

# Convert labels to numbers
df['species'] = df['species'].astype('category').cat.codes

# Train-test split
data = df.sample(frac=1, random_state=1).reset_index(drop=True)
X_train = data.iloc[:120, :-1].values
y_train = data.iloc[:120, -1].values
X_test  = data.iloc[120:, :-1].values
y_test  = data.iloc[120:, -1].values

# 2. Distance function
def dist(a, b):
    return np.sqrt(np.sum((a - b)**2))

# 3. Predict one sample
def knn_predict(sample, k=3):
    distances = []
    for i in range(len(X_train)):
        d = dist(sample, X_train[i])
        distances.append((d, y_train[i]))

    distances.sort()                  # sort by distance
    k_labels = [label for d, label in distances[:k]]
    return max(set(k_labels), key=k_labels.count)   # majority vote

# 4. Predict all
predictions = [knn_predict(x, k=5) for x in X_test]

# 5. Accuracy
accuracy = np.mean(predictions == y_test)
print("Accuracy:", accuracy)
```
